# Import Libraries

```python
from bs4 import BeautifulSoup
import requests
```

# Send 'get' requests to the webpage server to get the source code of the page

```python
page = requests.get('https://www.dineout.co.in/hyderabad-restaurants/trending')
```

```python
page
```

# Page content

```python
soup = BeautifulSoup(page.content)
soup
```

# Scraping First Name

```python
first_title = soup.find('div', class_="restnt-info cursor")
```

```python
first_title
```

```python
first_title.text
```

# Sraping First Location

```python
locate = soup.find('div', class_='restnt-loc ellipsis')

locate
```

```python
locate.text
```

# Scraping first Price

```python
sta = soup.find('span', class_="double-line-ellipsis")

sta.text
```

```python
sta.text.replace("₹","")
```

# Scraping Multiple Titles

```python
titles = []

for i in soup.find_all('div', class_="restnt-info cursor"):
    titles.append(i.text)
    
titles
```

# Scraping Multiple Locations

```python
locations = []

for i in soup.find_all('div', class_="restnt-loc ellipsis"):
    locations.append(i.text)
    
locations
```

# Scraping Multiple Prices

```python
prices = []

for i in soup.find_all('span', class_="double-line-ellipsis"):
    prices.append(i.text)
    
prices
```

# Scraping Multiple Images Urls

```python
images = []

for i in soup.find_all("img", class_='no-img'):
    images.append(i['data-src'])

images
```

```python
print(len(titles), len(locations), len(prices), len(images))
```

```python
import pandas as pd
df = pd.DataFrame({'Titles':titles, 'Locations':locations, 'Prices':prices, 'Images_url':images})
df
```
