```python
#importing Selenium, pandas, selenium webdriver, warnings, and time
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
#opening chrome browser using selenium driver
flipkart_driver = webdriver.Chrome()
```

```python
#opening the required webpage www.flipkart.com on automated chrome browser
flipkart_driver.get('https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART')
```

```python
#Declaring the lists
iphone11_rating = []
iphone11_review_summary = []
iphone11_full_review = []
```

```python
#scraping the data for iphone rating, reviews and full review
start = 0
end = 10
for page in range(start,end):
    ratings = flipkart_driver.find_elements(By.XPATH, '//div[@class="_3LWZlK _1BLPMq"]')
    for i in ratings:
        iphone11_rating.append(i.text)
    
    review_summary = flipkart_driver.find_elements(By.XPATH, '//p[@class="_2-N8zT"]')
    for i in review_summary:
        iphone11_review_summary.append(i.text)
        
    full_review = flipkart_driver.find_elements(By.XPATH, '//div[@class="t-ZTKy"]/div/div')
    for i in full_review:
        iphone11_full_review.append(i.text)
    
    next_button = flipkart_driver.find_element(By.XPATH, '//a[@class="_1LKTO3"]')
    next_button.click()
    time.sleep(10)
```

```python
print(len(iphone11_rating),len(iphone11_review_summary),len(iphone11_full_review))
```

Q4: Scrape data for first 100 sneakers you find when you visit flipkart.com and search for “sneakers” in the search field.
You have to scrape 3 attributes of each sneaker:
1. Brand
2. Product Description
3. Price

```python
#importing Selenium, pandas, selenium webdriver, warnings, and time
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
#opening chrome browser using selenium driver
flipkart_driver = webdriver.Chrome()
```

```python
#opening the required webpage www.flipkart.com on automated chrome browser
flipkart_driver.get('https://www.flipkart.com')
```

```python
#Enter 'sneakers' in the search field
flipkart_search_for = flipkart_driver.find_element(By.CLASS_NAME,"Pke_EE")
flipkart_search_for.send_keys('sneakers')
```

```python
#Clicking search button
flipkart_search = flipkart_driver.find_element(By.XPATH,'//div[@class="_1sFryS _2alaMB"]/button')
flipkart_search.click()
```

```python
#Declaring the lists
sneakers_brand = []
sneakers_description = []
sneakers_price = []
```

```python
#scraping the data for sneakers brand, product description and price
start = 0
end = 3
for page in range(start,end):
    brand = flipkart_driver.find_elements(By.CLASS_NAME, "_2WkVRV")
    for i in brand:
        sneakers_brand.append(i.text)
    
    description = flipkart_driver.find_elements(By.XPATH, '//div[@class="_2B099V"]/a[1]')
    for i in description:
        sneakers_description.append(i.text)
        
    price = flipkart_driver.find_elements(By.XPATH, '//div[@class="_2B099V"]/a[2]/div/div[1]')
    for i in price:
        sneakers_price.append(i.text)
    
    next_button = flipkart_driver.find_element(By.XPATH, '//a[@class="_1LKTO3"]')
    next_button.click()
    time.sleep(3)
```

```python
print(len(sneakers_brand),len(sneakers_description),len(sneakers_price))
```

```python
#creating dataframe
import pandas as pd
sneakers_df = pd.DataFrame({'Sneakers Brand':sneakers_brand[0:100],'Product Description':sneakers_description[0:100],'Price':sneakers_price[0:100]})
sneakers_df
```

scrape the mentioned data and make the DataFrame.

```python
#importing Selenium, pandas, selenium webdriver, warnings, and time
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
#opening chrome browser using selenium driver
josh_driver = webdriver.Chrome()
```

```python
#opening the webpage www.motor1.com on automated chrome browser
josh_driver.get("https://www.jagranjosh.com/general-knowledge/list-of-all-prime-ministers-of-india-1473165149-1")
```

```python
#Decalre Lists
prime_minister_name = []
pm_born_dead = []
pm_office_term = []
pm_remarks = []
```

```python
#Scraping data of prime ministers of india
pm_name_tags = josh_driver.find_elements(By.XPATH, '//div[@class="TableData"]/table/tbody/tr/td[2]/div')
for i in pm_name_tags:
    prime_minister_name.append(i.text)
    
pm_born_dead_tags = josh_driver.find_elements(By.XPATH, '//div[@class="TableData"]/table/tbody/tr/td[3]/div')
for i in pm_born_dead_tags:
    pm_born_dead.append(i.text)

pm_office_term_tags = josh_driver.find_elements(By.XPATH, '//div[@class="TableData"]/table/tbody/tr/td[4]')
for i in pm_office_term_tags:
    pm_term = i.text.replace("\n","::")
    pm_office_term.append(pm_term)
    
pm_remarks_tags = josh_driver.find_elements(By.XPATH, '//div[@class="TableData"]/table/tbody/tr/td[5]/div')
for i in pm_remarks_tags:
    pm_remarks.append(i.text)
    
print(len(prime_minister_name), len(pm_born_dead), len(pm_office_term), len(pm_remarks))
```

```python
#creating dataframe
import pandas as pd
pm_df = pd.DataFrame({'Prime Minister Name':prime_minister_name, 'Born-Dead':pm_born_dead, 'Office Term':pm_office_term, 'Remarks':pm_remarks})
print('List of Prime Ministers of India')
pm_df
```
