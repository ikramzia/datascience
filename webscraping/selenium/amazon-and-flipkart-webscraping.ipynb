Q1. Write a python program which searches all the product under a particular product from www.amazon.in. The product to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for guitars.

```python
#importing Selenium, pandas, selenium webdriver, warnings, and time
import selenium
import pandas as pd
from bs4 import BeautifulSoup
import time

#importing selenium webdriver
from selenium import webdriver

#importing warnings
import warnings
warnings.filterwarnings('ignore')

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

#Importing Exceptions
from selenium.common.exceptions import NoSuchElementException as NSEE
from selenium.common.exceptions import StaleElementReferenceException as SERE
from selenium.common.exceptions import ElementNotInteractableException as ENIE
from selenium.common.exceptions import TimeoutException as TE
from selenium.common.exceptions import InvalidSelectorException as ISE
```

```python
amazon_driver = webdriver.Chrome() #opening automated chrome browser using selenium driver
amazon_driver.maximize_window() #maximize window
amazon_driver.get('https://www.amazon.in') #opening the webpage www.amazon.com on automated chrome browser
time.sleep(2) #asking driver to wait for 2 seconds

search_product = input("Enter any product to be searched on Amazon!\n") #taking user input to search for on amazon

#searching for required product by the user and exception handling
try:
    amazon_product = amazon_driver.find_element(By.XPATH,'//div[@class="nav-search-field "]/input')
    amazon_product.send_keys(search_product)
except NSEE:
    print("Exception Raised: \n", NSEE)
except SERE:
    print("Exception Raised: \n", SERE)
except ENIE:
    print("Exception Raised: \n", ENIE)
except TE:
    print("Exception Raised: \n", TE)
except ISE:
    print("Exception Raised: \n", ISE)

try:
    amazon_search = amazon_driver.find_element(By.XPATH,'//span[@class="nav-search-submit-text nav-sprite nav-progressive-attribute"]/input')
    amazon_search.click()
except NSEE:
    print("Exception Raised: \n", NSEE)
except SERE:
    print("Exception Raised: \n", SERE)
except ENIE:
    print("Exception Raised: \n", ENIE)
except TE:
    print("Exception Raised: \n", TE)
except ISE:
    print("Exception Raised: \n", ISE)
    
try:
    search_result = amazon_driver.find_element(By.XPATH,'//div[@class="a-section a-spacing-small a-spacing-top-small"]')
    print('Got %s'%search_result.text)
except NSEE:
    print("Exception Raised: \n", NSEE)
except SERE:
    print("Exception Raised: \n", SERE)
except ENIE:
    print("Exception Raised: \n", ENIE)
except TE:
    print("Exception Raised: \n", TE)
except ISE:
    print("Exception Raised: \n", ISE)
```

3. Write a python program to access the search bar and search button on images.google.com and scrape 10 images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’, ‘Guitar’, ‘Cakes’.

```python
#importing Selenium, pandas, selenium webdriver, warnings, requests and time
import selenium
import pandas as pd
from bs4 import BeautifulSoup
import time
import requests

#importing selenium webdriver
from selenium import webdriver

#importing warnings
import warnings
warnings.filterwarnings('ignore')

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

#Importing Exceptions
from selenium.common.exceptions import NoSuchElementException as NSEE
from selenium.common.exceptions import StaleElementReferenceException as SERE
from selenium.common.exceptions import ElementNotInteractableException as ENIE
from selenium.common.exceptions import TimeoutException as TE
from selenium.common.exceptions import InvalidSelectorException as ISE
```

```python
google_driver = webdriver.Chrome() #opening automated chrome browser using selenium driver
google_driver.maximize_window() #maximize window
google_driver.get('https://images.google.com') #opening the webpage images.google.com on automated chrome browser
time.sleep(2) #asking driver to wait for 2 seconds

search_product = 'fruits' #to search for 'fruits' on google images

#searching for 'fruits' and exception handling
try:
    google_product = google_driver.find_element(By.XPATH,'//div[@class="a4bIc"]/textarea')
    google_product.clear()
    google_product.send_keys(search_product)
except NSEE:
    print("Exception Raised: \n", NSEE)
except SERE:
    print("Exception Raised: \n", SERE)
except ENIE:
    print("Exception Raised: \n", ENIE)
except TE:
    print("Exception Raised: \n", TE)
except ISE:
    print("Exception Raised: \n", ISE)

try:
    google_search_button = google_driver.find_element(By.XPATH,'//div[@class="RNNXgb"]/button')
    google_search_button.click()
except NSEE:
    print("Exception Raised: \n", NSEE)
except SERE:
    print("Exception Raised: \n", SERE)
except ENIE:
    print("Exception Raised: \n", ENIE)
except TE:
    print("Exception Raised: \n", TE)
except ISE:
    print("Exception Raised: \n", ISE)

time.sleep(5)
#scrolling webpage to get image results of upto 500
for _ in range(20):
    google_driver.execute_script("window.scrollBy(0,500)")

time.sleep(2)
fruit_img_urls = []  #declaring list
#scraping for image source
fruit_images = google_driver.find_elements(By.XPATH, '//img[@class="rg_i Q4LuWd"]')
for image in fruit_images:
    source = image.get_attribute('src')
    if source is not None:
        if(source[0:4] == 'http'):
            fruit_img_urls.append(source)

time.sleep(2)
#downloading 10 images from results
for i in range(len(fruit_img_urls)):
    if i > 10:
        break
    print("Downloading {0} of {1} images" .format(i, 10))
    response = requests.get(fruit_img_urls[i])
    file = open(r"C:\Users\ziaik\fruit_img"+str(i)+".jpg","wb")
    file.write(response.content)
```

```python
#importing Selenium, pandas, selenium webdriver, warnings, and time
import selenium
import pandas as pd
from bs4 import BeautifulSoup
import time

#importing selenium webdriver
from selenium import webdriver

#importing warnings
import warnings
warnings.filterwarnings('ignore')

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

#Importing Exceptions
from selenium.common.exceptions import NoSuchElementException as NSEE
from selenium.common.exceptions import StaleElementReferenceException as SERE
from selenium.common.exceptions import ElementNotInteractableException as ENIE
from selenium.common.exceptions import TimeoutException as TE
from selenium.common.exceptions import InvalidSelectorException as ISE
```

```python
flipkart_driver = webdriver.Chrome() #opening automated chrome browser using selenium driver
flipkart_driver.maximize_window() #maximize window
flipkart_driver.get('https://www.flipkart.com') #opening the webpage www.flipkart.com on automated chrome browser
time.sleep(2) #asking driver to wait for 2 seconds

search_product = 'Oneplus Nord'

#searching for smartphone and exception handling
try:
    flipkart_product = flipkart_driver.find_element(By.XPATH,'//input[@class="Pke_EE"]')
    flipkart_product.send_keys(search_product)
except NSEE:
    print("Exception Raised: \n", NSEE)
except SERE:
    print("Exception Raised: \n", SERE)
except ENIE:
    print("Exception Raised: \n", ENIE)
except TE:
    print("Exception Raised: \n", TE)
except ISE:
    print("Exception Raised: \n", ISE)

try:
    flipkart_search = flipkart_driver.find_element(By.XPATH,'//button[@class="_2iLD__"]')
    flipkart_search.click()
except NSEE:
    print("Exception Raised: \n", NSEE)
except SERE:
    print("Exception Raised: \n", SERE)
except ENIE:
    print("Exception Raised: \n", ENIE)
except TE:
    print("Exception Raised: \n", TE)
except ISE:
    print("Exception Raised: \n", ISE)
    
try:
    search_result = flipkart_driver.find_element(By.XPATH,'//span[@class="_10Ermr"]')
    print(search_result.text.replace("Showing", "Got"))
except NSEE:
    print("Exception Raised: \n", NSEE)
except SERE:
    print("Exception Raised: \n", SERE)
except ENIE:
    print("Exception Raised: \n", ENIE)
except TE:
    print("Exception Raised: \n", TE)
except ISE:
    print("Exception Raised: \n", ISE)
```

```python
#importing Selenium, pandas, selenium webdriver, warnings, and time
import selenium
import pandas as pd
from bs4 import BeautifulSoup
import time

#importing selenium webdriver
from selenium import webdriver

#importing warnings
import warnings
warnings.filterwarnings('ignore')

from time import sleep
from selenium.webdriver import ActionChains
from selenium.webdriver.common.actions.action_builder import ActionBuilder
from selenium.webdriver.common.actions.mouse_button import MouseButton
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

#Importing Exceptions
from selenium.common.exceptions import NoSuchElementException as NSEE
from selenium.common.exceptions import StaleElementReferenceException as SERE
from selenium.common.exceptions import ElementNotInteractableException as ENIE
from selenium.common.exceptions import TimeoutException as TE
from selenium.common.exceptions import InvalidSelectorException as ISE
from selenium.common.exceptions import ElementClickInterceptedException as ECIE
```

```python
hostelworld_driver = webdriver.Chrome() #opening automated chrome browser using selenium driver
hostelworld_driver.maximize_window() #maximize window
hostelworld_driver.get('https://www.hostelworld.com/') #opening the video webpage www.hostelworld.com on automated chrome browser

time.sleep(2) #asking driver to wait for 2 seconds

#searching for hostels in 'London' as location and exception handling
location = 'London'

try:
    hotels_location = hostelworld_driver.find_element(By.XPATH,'//div[@class="search-form-wrapper"]/div/div/div[1]/div[1]/div/div[2]/input')
    hotels_location.send_keys(location)
except NSEE:
    print("Exception Raised: \n", NSEE)
except SERE:
    print("Exception Raised: \n", SERE)
except ENIE:
    print("Exception Raised: \n", ENIE)
except TE:
    print("Exception Raised: \n", TE)
except ISE:
    print("Exception Raised: \n", ISE)
except ECIE:
    print("Exception Raised: \n", ECIE)

time.sleep(3)

try:
    select_city = hostelworld_driver.find_element(By.XPATH,'//div[@class="destination-selection"]/div/ul/li[2]')
    select_city.click()
except NSEE:
    print("Exception Raised: \n", NSEE)
except SERE:
    print("Exception Raised: \n", SERE)
except ENIE:
    print("Exception Raised: \n", ENIE)
except TE:
    print("Exception Raised: \n", TE)
except ISE:
    print("Exception Raised: \n", ISE)
except ECIE:
    print("Exception Raised: \n", ECIE)

time.sleep(2)

try:
    search_button = hostelworld_driver.find_element(By.XPATH,'//button[@class="btn-content large-button"]')
    search_button.click()
except NSEE:
    print("Exception Raised: \n", NSEE)
except SERE:
    print("Exception Raised: \n", SERE)
except ENIE:
    print("Exception Raised: \n", ENIE)
except TE:
    print("Exception Raised: \n", TE)
except ISE:
    print("Exception Raised: \n", ISE)
except ECIE:
    print("Exception Raised: \n", ECIE)

time.sleep(10)

try:
    search_result = hostelworld_driver.find_element(By.XPATH,'//h4[@class="body-1-reg"]')
    print(search_result.text.replace("Showing", "Got"),'in %s'%location)
except NSEE:
    print("Exception Raised: \n", NSEE)
except SERE:
    print("Exception Raised: \n", SERE)
except ENIE:
    print("Exception Raised: \n", ENIE)
except TE:
    print("Exception Raised: \n", TE)
except ISE:
    print("Exception Raised: \n", ISE)
except ECIE:
    print("Exception Raised: \n", ECIE)
```

```python
#creating lists
hostel_distance = []
privates_from_price = []
dorms_from_price = []
hostelworld_url = []

#Next page button always giving ElementClickIntercepedException 

#getting page url & iterating to next pages
url = hostelworld_driver.current_url
hostel_result_url = url.replace("page=1","page=")
for page in range(1, 5):
    try:
        hostelworld_page = hostel_result_url + str(page)
        hostelworld_driver.get(hostelworld_page)
        time.sleep(5)
        
        #scraping 'hostel distance from city centre' from all pages and exception handling
        distance = hostelworld_driver.find_elements(By.XPATH, '//span[@class="distance-description"]')
        for i in distance:
            try:
                hostel_distance.append(i.text.replace('- ',""))
            except NSEE:
                hostel_distance.append('--')
            except SERE:
                hostel_distance.append('--')
            except ENIE:
                hostel_distance.append('--')
            except TE:
                hostel_distance.append('--')
            except ISE:
                hostel_distance.append('--')
            except ECIE:
                hostel_distance.append('--')

        #scraping 'privates from price' from next pages and exception handling
        privates_price = hostelworld_driver.find_elements(By.XPATH, '//div[@class="property-accommodations"]/div/div[1]')
        for i in privates_price:
            try:
                privates_from_price.append(i.text.split()[-1].replace("Available", "No Privates Available"))
            except NSEE:
                privates_from_price.append('--')
            except SERE:
                privates_from_price.append('--')
            except ENIE:
                privates_from_price.append('--')
            except TE:
                privates_from_price.append('--')
            except ISE:
                privates_from_price.append('--')
            except ECIE:
                privates_from_price.append('--')

        #scraping 'dorms from price' from all pages and exception handling
        dorms_price = hostelworld_driver.find_elements(By.XPATH, '//div[@class="property-accommodations"]/div/div[2]')
        for i in dorms_price:
            try:
                dorms_from_price.append(i.text.split()[-1].replace("Available", "No Dorms Available"))
            except NSEE:
                dorms_from_price.append('--')
            except SERE:
                dorms_from_price.append('--')
            except ENIE:
                dorms_from_price.append('--')
            except TE:
                dorms_from_price.append('--')
            except ISE:
                dorms_from_price.append('--')
            except ECIE:
                dorms_from_price.append('--')

        #scraping 'hostel url' from all pages and exception handling
        hostel_url = hostelworld_driver.find_elements(By.XPATH, '//a[@class="property-card-container horizontal"]')
        for i in hostel_url:
            try:
                hostelworld_url.append(i.get_attribute('href'))
            except NSEE:
                hostelworld_url.append('--')
            except SERE:
                hostelworld_url.append('--')
            except ENIE:
                hostelworld_url.append('--')
            except TE:
                hostelworld_url.append('--')
            except ISE:
                hostelworld_url.append('--')
            except ECIE:
                hostelworld_url.append('--')
    except:
        print('URL not found!')
```

```python
#creating lists
hostel_name = []
hostel_rating = []
total_reviews = []
overall_review = []
hostel_facilities = []
property_description = []

for i in hostelworld_url: #iterating each and every url to fetch all hostel details 
    hostelworld_driver.get(i)
    time.sleep(10)
    
    #scraping for hostel name
    try:
        name = hostelworld_driver.find_element(By.XPATH, '//div[@class="title-2"]')
        hostel_name.append(name.text)
    except NSEE:
        hostel_name.append('--')
    except SERE:
        hostel_name.append('--')
    except ENIE:
        hostel_name.append('--')
    except TE:
        hostel_name.append('--')
    except ISE:
        hostel_name.append('--')
    except ECIE:
        hostel_name.append('--')

    #scraping for hostel rating
    try:
        rating = hostelworld_driver.find_element(By.XPATH, '//div[@class="reviews-and-ratings"]/div[1]/div/div[1]')
        hostel_rating.append(rating.text)
    except NSEE:
        hostel_rating.append('--')
    except SERE:
        hostel_rating.append('--')
    except ENIE:
        hostel_rating.append('--')
    except TE:
        hostel_rating.append('--')
    except ISE:
        hostel_rating.append('--')
    except ECIE:
        hostel_rating.append('--')

    #scraping for hostel total reviews
    try:
        t_reviews = hostelworld_driver.find_element(By.XPATH, '//div[@class="reviews-and-ratings"]/div[1]/div/div[2]/div[2]')
        total_reviews.append(t_reviews.text.split()[0]+" "+t_reviews.text.split()[2])
    except NSEE:
        total_reviews.append('--')
    except SERE:
        total_reviews.append('--')
    except ENIE:
        total_reviews.append('--')
    except TE:
        total_reviews.append('--')
    except ISE:
        total_reviews.append('--')
    except ECIE:
        total_reviews.append('--')

    #scraping for hostel Overall review
    try:
        o_review = hostelworld_driver.find_element(By.XPATH, '//div[@class="reviews-and-ratings"]/div[1]/div/div[2]/div[1]')
        overall_review.append(o_review.text)
    except NSEE:
        overall_review.append('--')
    except SERE:
        overall_review.append('--')
    except ENIE:
        overall_review.append('--')
    except TE:
        overall_review.append('--')
    except ISE:
        overall_review.append('--')
    except ECIE:
        overall_review.append('--')

    #scraping for hostel facilities
    try:
        facility_1 = hostelworld_driver.find_element(By.XPATH, '//ul[@class="facility-sections"]/li[1]/ul/li[1]')
        facility_2 = hostelworld_driver.find_element(By.XPATH, '//ul[@class="facility-sections"]/li[1]/ul/li[2]')
        facility_3 = hostelworld_driver.find_element(By.XPATH, '//ul[@class="facility-sections"]/li[1]/ul/li[3]')
        facility_4 = hostelworld_driver.find_element(By.XPATH, '//ul[@class="facility-sections"]/li[1]/ul/li[4]')
        facilities = facility_1.text+", "+facility_2.text+", "+facility_3.text+", "+facility_4.text
        hostel_facilities.append(facilities)
    except NSEE:
        hostel_facilities.append('--')
    except SERE:
        hostel_facilities.append('--')
    except ENIE:
        hostel_facilities.append('--')
    except TE:
        hostel_facilities.append('--')
    except ISE:
        hostel_facilities.append('--')
    except ECIE:
        hostel_facilities.append('--')
        

    #scraping for property description
    try:
        description = hostelworld_driver.find_element(By.XPATH, '//div[@class="flex-80"]/div[6]/div/div[2]')
        property_description.append(description.text.replace("\n",""))
    except NSEE:
        property_description.append('--')
    except SERE:
        property_description.append('--')
    except ENIE:
        property_description.append('--')
    except TE:
        property_description.append('--')
    except ISE:
        property_description.append('--')
    except ECIE:
        property_description.append('--')
```
