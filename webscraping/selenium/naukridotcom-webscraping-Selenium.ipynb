```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
driver = webdriver.Chrome()
```

```python
driver.get('https://www.naukri.com/')
```

```python
designation = driver.find_element(By.CLASS_NAME,"suggestor-input ")
designation.send_keys('Data Analyst')
```

```python
location = driver.find_element(By.XPATH,"/html/body/div[1]/div[7]/div/div/div[5]/div/div/div/div[1]/div/input")
location.send_keys('Bangalore')
```

```python
search = driver.find_element(By.CLASS_NAME,"qsbSubmit")
search.click()
```

```python
job_title = []
job_location = []
job_company = []
job_experience = []
```

```python
title_tags = driver.find_elements(By.XPATH,'//div[@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]/div/a')
for i in title_tags:
    title = i.text
    job_title.append(title)
    
location_tags = driver.find_elements(By.XPATH,'//span[@class="locWdth"]')
for i in location_tags:
    location = i.text
    job_location.append(location)
    
#company_tags = driver.find_elements(By.XPATH,'//div[@class=" row2"]/span/a[1]')
#rating_tags = driver.find_elements(By.XPATH,'//div[@class=" row2"]/span/a[2]')
#review_tags = driver.find_elements(By.XPATH,'//div[@class=" row2"]/span/a[3]')
#company_tags = driver.find_elements(By.XPATH,'//a[@class=" comp-name mw-25"]')
#for i in company_tags:
#    company = i.text
#    job_company.append(company)
    
experience_tags = driver.find_elements(By.XPATH,'//span[@class="expwdth"]')
for i in experience_tags:
    experience = i.text
    job_experience.append(experience)
```

```python
start = 0
end = 3
for page in range(start,end):
    titles = driver.find_elements(By.XPATH, '//div[@class="cust-job-tuple layout-wrapper lay-2 sjw__tuple "]/div/a')
    for i in titles:
        job_titles.append(i.text)
    next_button = driver.find_element(By.XPATH, '/html/body/div/div/main/div[1]/div[2]/div[3]/div/a[2]')
    next_button.click()
    time.sleep(3)
```
